# Convert Tables
Copy tables from One Note into notebook and format for analysis

```python
### import packages
import pandas as pd
```

## Convert

Copy table from One Note then run cells below

```python
# read table from clipboard
df_clip=pd.read_clipboard()
# display dataframe
display(df_clip)
```

### Formatting

Edit dataframe
 - rename columns
 - split x/y data per column

```python
### make copy and rename columns
df_data=df_clip.rename(columns={df_clip.columns[0]: "particle_type", 
                                df_clip.columns[1]: "entries",
                                df_clip.columns[2]: "mean",
                                df_clip.columns[3]: "std_dev",
                                })
### split X/Y data
# copy original dataframe
df_x=df_data.copy(deep=True)
df_x['axis']="X"
df_y=df_data.copy(deep=True)
df_y['axis']="Y"
# define data columns
data_columns=['entries','mean','std_dev']
# loop over data columns
for col in data_columns:
    # split data: x is first (0th element)
    df_x[col]=df_x[col].apply(lambda l: l.split('/')[0])
    # split data: y is second (1th element)
    df_y[col]=df_y[col].apply(lambda l: l.split('/')[1])
# combine dataframes
df_data=pd.concat([df_x,df_y]).reset_index(drop=True)
### get energy from particle_type column
df_data['energy']=df_data['particle_type'].apply(lambda l: l.split('(')[1].split(')')[0])
# tidy particle_type column
df_data['particle_type']=df_data['particle_type'].apply(lambda l: l.split('(')[0].strip())
### have a look
display(df_data)
```

### write to csv

write data to csv, then copy to other notebook

```python
df_data.to_csv(index=False)
```
